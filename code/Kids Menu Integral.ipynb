```python
import numpy as np
import matplotlib

import matplotlib.pyplot as plt
```

```python
t = 1.0

N = 10000

dt = t / N
```

```python
dts = [dt] * N

dts[:20]
```

```python
ts = [i*dt for i in range(N)]
```

```python
dts[:20]
```

```python
ts_tds = [x*y for (x,y) in zip(ts, dts)]
```

```python
plt.plot(ts, ts_tds)
```

```python
sum(ts_tds)
```

```python
dts_mod = [x for x in dts]

dts_mod[10] = 3*dt / 2
dts_mod[128] = dt / 2
```

```python
sum([x*y for (x,y) in zip(ts, dts_mod)])
```
