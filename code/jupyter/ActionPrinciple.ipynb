```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
h = 1.0
g = 9.8

T = np.sqrt(2*h/g)
```

```python
def q1(t):
    y = h - g * t * t / 2
    return y

def q2(t):
    y = h * (1 - t / T)

    return y

def q3(t):
    y = h * (1 - np.sin(np.pi*t/2/T))
    return y

def q4(t):
    y = h * (np.exp(1)*np.exp(-t/T)-1)/(np.exp(1)-1)

    return y

def q5(t):
    y = h * np.power((1 - t / T), 2)

    return y
```

```python
ts = np.linspace(0, T, 100)

y1s = q1(ts)
y2s = q2(ts)
y3s = q3(ts)
y4s = q4(ts)
y5s = q5(ts)
```

```python
fig = plt.plot(figsize=[8 ,6])
alpha = 0.6

plt.plot(ts, y1s, label="1 - Free fall", alpha=alpha)
plt.plot(ts, y2s, label="2 - Const speed", alpha=alpha)
plt.plot(ts, y3s, label="3 - Harmonic", alpha=alpha)
plt.plot(ts, y4s, label="4 - Exponential", alpha=alpha)
plt.plot(ts, y5s, label="5 - Inv parabola", alpha=alpha)

plt.legend(loc="upper right")
plt.xlabel("Time (s)")
plt.ylabel("Y (m)")

plt.grid(1)

plt.show()
plt.close()
```
